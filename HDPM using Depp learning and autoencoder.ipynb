```python
import pandas as pd
from imblearn.over_sampling import RandomOverSampler
from imblearn.combine import SMOTEENN

# Separate the features and target variable
X_train = df_selected.drop("HeartDiseaseorAttack", axis=1)
y_train = df_selected["HeartDiseaseorAttack"]

# Perform SMOTEENN oversampling on the resampled data
smote_enn = SMOTEENN(random_state=42)
X_train_resampled_smoteenn, y_train_resampled_smoteenn = smote_enn.fit_resample(X_train, y_train)

# Perform RandomOverSampler on the training data
random_oversampler = RandomOverSampler(random_state=42)
X_train_resampled, y_train_resampled = random_oversampler.fit_resample(X_train_resampled_smoteenn, y_train_resampled_smoteenn)

# Combine the resampled features and target variable into a single DataFrame
resampled_data = pd.concat([X_train_resampled, y_train_resampled], axis=1)

# Separate the instances with 0 and 1 values
data_0 = resampled_data[resampled_data["HeartDiseaseorAttack"] == 0]
data_1 = resampled_data[resampled_data["HeartDiseaseorAttack"] == 1]

# Sample 3000 instances from each class
data_0_sample = data_0.sample(n=3000, random_state=42)
data_1_sample = data_1.sample(n=3000, random_state=42)

# Concatenate the sampled instances into a single DataFrame
first_csv_data = pd.concat([data_0_sample, data_1_sample])

# Drop the sampled instances from the original data to get the remaining instances
remaining_data_0 = data_0.drop(data_0_sample.index)
remaining_data_1 = data_1.drop(data_1_sample.index)
second_csv_data = pd.concat([remaining_data_0, remaining_data_1])

# Save the DataFrames to CSV files
first_csv_data.to_csv("test.csv", index=False)
second_csv_data.to_csv("heart_disease_train_resampled.csv", index=False)
```

```python
#====== Updated
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input

# Load the dataset
df = pd.read_csv("heart_disease_train_resampled.csv")

# Separate the features and target variable
X_train = df.drop("HeartDiseaseorAttack", axis=1)
y_train = df["HeartDiseaseorAttack"]

# Normalize the data
scaler = MinMaxScaler()
X_train_normalized = scaler.fit_transform(X_train)

# Define the autoencoder architecture
input_shape = X_train_normalized.shape[1]
encoding_dim = 64  # Adjust the size of the encoding layer as per your choice

# Encoder
input_layer = Input(shape=(input_shape,))
encoder = Dense(encoding_dim, activation='relu')(input_layer)

# Decoder
decoder = Dense(input_shape, activation='sigmoid')(encoder)

# Autoencoder
autoencoder = Model(inputs=input_layer, outputs=decoder)

# Compile and train the autoencoder
autoencoder.compile(optimizer='adam', loss='mse')
autoencoder.fit(X_train_normalized, X_train_normalized, epochs=10, batch_size=32, validation_split=0.2)

# Save the trained model
autoencoder.save("autoencoder.h5")
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.optimizers import Adam

# Load the training dataset
df_train = pd.read_csv("heart_disease_train.csv")

# Load the validation dataset
df_val = pd.read_csv("heart_disease_validation.csv")

# Separate the features and target variable for training
X_train = df_train.drop("HeartDiseaseorAttack", axis=1)
y_train = df_train["HeartDiseaseorAttack"]

# Separate the features and target variable for validation
X_val = df_val.drop("HeartDiseaseorAttack", axis=1)
y_val = df_val["HeartDiseaseorAttack"]

# Normalize the training data
scaler = MinMaxScaler()
X_train_normalized = scaler.fit_transform(X_train)

# Normalize the validation data
X_val_normalized = scaler.transform(X_val)


# Obtain the extracted features
extracted_features = autoencoder.predict(X_train_normalized)
extracted_features_val = autoencoder.predict(X_val_normalized)

# Define the RNN model
rnn_model = Sequential()
rnn_model.add(LSTM(128, input_shape=(extracted_features.shape[1], 1), return_sequences=True))
rnn_model.add(Dropout(0.2))
rnn_model.add(LSTM(64))
rnn_model.add(Dropout(0.2))
rnn_model.add(Dense(1, activation='sigmoid'))

# Compile the rnn_model with a custom learning rate
optimizer = Adam(lr=0.001)
rnn_model.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=['accuracy'])

# Train the model
rnn_model.fit(
    np.expand_dims(extracted_features, axis=2),
    y_train,
    epochs=20,
    batch_size=32,
    validation_data=(np.expand_dims(extracted_features_val, axis=2), y_val)
)

# Save the trained model
rnn_model.save("rnn_model.h5")
```

```python
#============== cross fold cnn 

import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, Flatten, Dense
from sklearn.model_selection import KFold

# Load the training dataset
df_train = pd.read_csv("heart_disease_train_resampled.csv")

# Separate the features and target variable
X = df_train.drop("HeartDiseaseorAttack", axis=1)
y = df_train["HeartDiseaseorAttack"]

# Normalize the data
scaler = MinMaxScaler()
X_normalized = scaler.fit_transform(X)

autoencoder = load_model("autoencoder.h5")

# Obtain the extracted features
extracted_features = autoencoder.predict(X_normalized)


# Reshape the data for CNN
X_2d = np.expand_dims(extracted_features, axis=2)
n_timesteps, n_features = X_2d.shape[1], X_2d.shape[2]

# CNN Initialization function
def create_cnn_model():
    model = Sequential()
    model.add(Conv1D(filters=8, kernel_size=2, activation="relu", input_shape=(n_timesteps, n_features)))
    model.add(MaxPooling1D(pool_size=2))
    model.add(Flatten())
    model.add(Dense(16, activation="relu"))
    model.add(Dense(1, activation="sigmoid"))
    return model

# Compile the model
def compile_model(model):
    model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
    return model

# Perform 5-fold cross-validation
num_folds = 5
kf = KFold(n_splits=num_folds, shuffle=True)

fold = 1
for train_index, val_index in kf.split(X_2d):
    print(f"Training fold {fold}...")
    
    X_train_fold, X_val_fold = X_2d[train_index], X_2d[val_index]
    y_train_fold, y_val_fold = y[train_index], y[val_index]
    
    cnn_model = create_cnn_model()
    cnn_model = compile_model(cnn_model)
    
    cnn_model.fit(
        X_train_fold,
        y_train_fold,
        epochs=10,
        batch_size=32,
        validation_data=(X_val_fold, y_val_fold)
    )
    
    fold += 1

# Save the trained CNN model (you can choose to save the best model based on cross-validation)
cnn_model.save("cnn_model_cf.h5")




```
